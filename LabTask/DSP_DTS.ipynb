### **Discrete Time System**

#### ***Task:*** *An discrete signal and an integer is given. Determine the i) Time Shifted ii) Folding and iii) Scaled signal.*

**Input:**
- Discrete Signal
- An integer


**Output:**
- Time Shifted Signal
- Folding Signal
- Scaled Signal

### **Time Shifting of a Discrete Signal**

```python
import numpy as np
import matplotlib.pyplot as plt
import copy as cpy

f = input("Enter the value of frequency: ")
fs = input("Enter the value of sampling frequency: ")

print(f)
print(fs)

Ts = float(fs)/float(f)
n = int(Ts)

t = np.linspace(0,float(1/float(f)),1000)
ts = np.linspace(0,float(1/float(f)),n)
tf = np.linspace(-1*float(1/float(f)),float(1/float(f)),n)

x = np.cos(2.0*np.pi*float(f)*ts)

plt.stem(ts,x)
```

```python
## Take an integer k and shift by k

k = input("Enter the value of K: ")
print("value of k: ",k)

kVal = ts[1]
y = np.cos(2.0*np.pi*float(f)*(ts+int(k)*kVal))

plt.stem(ts,y)

```

### **Folding of a Discrete Signal**

```python
## Folding of a Signal

#z = 2*np.cos(2.0*np.pi*float(f)*tf)
print(tf)

z = 2*tf*tf- 30*tf - 1
print(z)

plt.stem(tf,z)
```

```python
n = [ -1*x for x in tf]
z  = [2*x*x- 30*x - 1 for x in n]

print(tf)

plt.stem(tf,z)

print(z)
```

### **Scaling of a Discrete Signal**

```python
f = input("Enter the value of frequency: ")
fs = input("Enter the value of sampling frequency: ")

print(f)
print(fs)

Ts = float(fs)/float(f)
n = int(Ts)

t = np.linspace(0,float(1/float(f)),1000)
ts = np.linspace(-1*float(1/float(f)),float(1/float(f)),2*n)

x = np.cos(2.0*np.pi*float(f)*ts)

plt.stem(ts,x)
```

```python
m = input("Enter the scaling multiplier: ")

M = 2*n/float(m)
N = int(M)

tf = np.linspace(-1*float(1/float(f)),float(1/float(f)),N)
y = np.cos(2.0*np.pi*float(f)*tf)


plt.stem(tf,y)
```
