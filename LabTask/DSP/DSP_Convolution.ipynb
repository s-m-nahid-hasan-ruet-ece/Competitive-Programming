```python
import numpy as np
import matplotlib.pyplot as plt


h=[]
x=[]

l = int(input("Enter the lower limit of h(n) time:  "))
r = int(input("Enter the upper limit of h(n) time:  "))

for i in range(int(l),int(r)+1):
    tval = input("Enter the value of h(n) for t = %d:  " %(i))
    h.append(int(tval))

print("Time Limit for h(n): [%d, %d]"%(int(l),int(r)))
print("h(n): ",h)

L = int(input("Enter the lower limit of x(n) time:  "))
R = int(input("Enter the upper limit of x(n) time:  "))

for i in range(int(L),int(R)+1):
    tval = input("Enter the value of x(n) for t = %d:  " %(i))
    x.append(int(tval))

print("Time Limit for x(n): [%d, %d]"%(int(L),int(R)))
print("x(n): ",x)

low = min(l,L)
high = low + (len(x)+len(h)-2)


ht = np.linspace(int(l),int(r),(int(r)-int(l)+1))
xt = np.linspace(int(L),int(R),(int(R)-int(L)+1))
ct = np.linspace(low,high,high-low+1)

convMap =  {}
conv=[]

for i in range(0,len(h)+len(x)):
    convMap[i] = 0

for i in range(0,len(h)):
    s = 1
    for j in range(0, len(x)):        
        convMap[i+j]+= (h[i]*x[j])
    

for i in range(0,len(h)+len(x)-1):
    conv.append(convMap[i])

print("y(n): ",conv)


plt.figure(figsize=(25, 8))
plt.title("Plotting of h(n)")
plt.stem(ht,h)

```

```python
plt.title("Plotting of x(n)")
plt.figure(figsize=(25, 8))
plt.stem(xt,x)
```

```python
plt.figure(figsize=(25, 8))
plt.stem(ct,conv)
```
