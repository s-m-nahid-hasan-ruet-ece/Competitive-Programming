### **Discrete Time System**

#### ***Task:*** *A discrete signal and an integer is given. Determine the i) Time Shifted ii) Folding and iii) Scaled signal.*

**Input:**
- Discrete Signal
- An integer


**Output:**
- Time Shifted Signal
- Folding Signal
- Scaled Signal

### **Time Shifting of a Discrete Signal**

```python
import numpy as np
import matplotlib.pyplot as plt
import copy as cpy

f = input("Enter the value of frequency: ")
fs = input("Enter the value of sampling frequency: ")

print(f)
print(fs)

Ts = float(fs)/float(f)
n = int(Ts)

ts = np.linspace(0,float(2/float(f)),2*n)
tf = np.linspace(-1*float(1/float(f)),float(1/float(f)),2*n)

x = np.cos(2.0*np.pi*float(f)*ts)

plt.figure(figsize=(25, 8))
plt.stem(ts,x)
```

```python
## Take an integer k and shift by k

k = input("Enter the value of K: ")
print("value of k: ",k)

kVal = ts[1]
y = np.cos(2.0*np.pi*float(f)*(ts+int(k)*kVal))

plt.figure(figsize=(25, 8))
plt.stem(ts,y)

```

### **Folding of a Discrete Signal**

```python
f = input("Enter the value of frequency: ")
fs = input("Enter the value of sampling frequency: ")

print(f)
print(fs)

Ts = float(fs)/float(f)
n = int(Ts)

ts = np.linspace(-1*float(1/float(f)),float(1/float(f)),2*n)

x = np.cos(2.0*np.pi*float(f)*ts)

plt.figure(figsize=(25, 8))
plt.stem(ts,x)
```

## **Time Variant**

```python
# x(n) -> T -> n * x(n) = y(n)

import numpy as np
import matplotlib.pyplot as plt
import copy as cpy

f = input("Enter the value of frequency: ")
fs = input("Enter the value of sampling frequency: ")

print(f)
print(fs)

Ts = float(fs)/float(f)
n = int(Ts)

ts = np.linspace(0,float(2/float(f)),2*n)
tf = np.linspace(-1*float(1/float(f)),float(1/float(f)),2*n)

x = np.cos(2.0*np.pi*float(f)*ts)

plt.figure(figsize=(25, 8))
plt.stem(ts,x)
```

```python
## passing through DTS
k = input("Enter the value of k: ")

kVal = ts[1]

kVal = ts[1]
ts = np.linspace(0,float(1/float(f)),1*n)

y = (ts+int(k)*kVal)*(ts+int(k)*kVal)* np.sin(2.0*np.pi*float(f)*(ts+int(k)*kVal))
yy = ts* ts* np.sin(2.0*np.pi*float(f)*(ts+int(k)*kVal))
plt.figure(figsize=(25, 8))
plt.stem(ts,y)

print(y)
print(yy)
```

```python
## yy!= y
plt.figure(figsize=(25, 8))
plt.stem(ts,yy)
```
