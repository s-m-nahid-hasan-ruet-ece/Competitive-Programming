### **Sampling of an Analog Signal**

#### ***Task:*** *An analog signal and sampling frequency is given. Determine the sampled signal.*

**Input:**
- Signal Quantity
- Sampling Frequency


**Output:**
- Output Signal
- Output Waveshape

```python
import numpy as np
import matplotlib.pyplot as plt
import copy as cpy

f = input("Enter the value of frequency: ")
fs = input("Enter the value of sampling frequency: ")

print(f)
print(fs)

Ts = float(fs)/float(f)
n = int(Ts)

t = np.linspace(0,float(1/float(f)),100)
ts = np.linspace(0,float(1/float(f)),n)

#t = np.arange(0,100,0.1)
#ts = np.arange(0,1000,10)

print(len(t))
x = np.sin(2.0*np.pi*float(f)*t)
y = np.sin(2.0*np.pi*float(f)*ts)
#plt.plot(t,x)
#y = np.sin((2.0 *np.pi*t)*(float(f)/float(fs)))
plt.plot(t,x)

```

```python
L = input("Enter the number of bits to transmit a sample: ")
print(L)
pMax = np.max(x)
nMax = np.min(x)

nLevel = (pMax-nMax)/pow(2,int(L))

print(nLevel)

xx = cpy.deepcopy(x)

vMax = pMax-nLevel/2
vMin = nMax+nLevel/2

#for i in range(vMin,vMax,nLevel):
for j in range(len(xx)):
    p = (int)(xx[j]/nLevel)
    if(abs(xx[j]-p*nLevel)<abs(((p+1)*nLevel)-xx[j])):
        xx[j] = p*nLevel          
    else:
        xx[j] = ((p+1)*nLevel)


plt.plot(t,xx)



```
