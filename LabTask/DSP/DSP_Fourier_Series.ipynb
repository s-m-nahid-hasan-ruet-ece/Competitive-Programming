### **Fourier Series**

```python
import numpy as np
import matplotlib.pyplot as plt


l = int(input("Enter the lower limit k:  "))
r = int(input("Enter the upper limit k:  "))

print("ak=[%d , %d]"%(l,r))

n = np.linspace(l,r,(r-l+1))

magnitude=[]
Phase=[]

for i in range(l,r+1):
    R = float(input("Enter the real part of a[%d]:  "%(i)))
    I = float(input("Enter the imaginary part of a[%d]:  "%(i)))
    z = complex(R,I)
    print("a[%d]: %f + %fj"%(i,R,I))
    magnitude.append(abs(z))
    Phase.append((180*np.angle(z))/np.pi)

plt.figure(figsize=(25, 8))
plt.title("Plotting of magnitude spectrum")
plt.stem(n,magnitude)
```

```python
plt.figure(figsize=(25, 8)) 
plt.title("Plotting of phase spectrum")
plt.stem(n,Phase)
```
