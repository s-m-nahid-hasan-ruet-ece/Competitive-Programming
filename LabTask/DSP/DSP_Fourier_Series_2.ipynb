```python
import numpy as np
import matplotlib.pyplot as plt

T = 20
nmax = 100
Ts = T/(2*nmax)
t = np.linspace(0,T-Ts,2*nmax)
print(t)
```

```python
A = 2
f=[]

for i in range(len(t)):
    if(t[i]< T/2):
        f.append(A)
    else:
        f.append(-1*A)
plt.figure(figsize=(25, 8)) 
plt.plot(t,f)
```

```python
n = 10

cosPart = np.cos(2*np.pi*t*n/T)
sinPart = np.sin(2*np.pi*t*n/T)

plt.plot(t,cosPart)
plt.plot(t,sinPart)
plt.figure(figsize=(25, 8)) 
plt.show()
```

### **Fourier Coefficients or Analysis with Real Coefficients**
Using sine and cosine functions it is possible to decompose a periodic signal into its harmonics. The coefficients are real valued. The analysis formula for real cofficients is as follows:

!["Image"](img.png)

```python
N = 30

ind = 1
a = []
b = []

for i in range(0,N):
    cosPart = np.cos(2*np.pi*t*i/T)
    sinPart = np.sin(2*np.pi*t*i/T)
    a.append( (2*Ts/T)*sum(np.array(f) * np.array(cosPart)))
    b.append( (2*Ts/T)*sum(np.array(f) * np.array(sinPart)))
    ind+=1

x = np.arange(N)
print(x)
plt.stem(x,a, 'b', markerfmt='bo', label='cos')
plt.stem(x,b, 'g', markerfmt='go', label='sin')
plt.legend(["an","bn"])
plt.figure(figsize=(25, 8)) 

plt.show()
    
```

### **Fourier Series or Synthesis with Real Coefficients**
We can now write the Fourier series representation of the function f(t), which is also known as synthesis. 
!["Image1"](img1.png)

```python
fs = []

for i in range(len(t)):
    fs.append(a[1]/2)


for i in range(1,N-1):
    cosPart = np.cos(2*np.pi*t*i/T)
    sinPart = np.sin(2*np.pi*t*i/T)
    fs = fs + a[i+1]*cosPart + b[i+1]*sinPart


plt.plot(t,f)
plt.plot(t,fs)
plt.show()


```

### **Fourier Coefficients or Analysis with Complex Coefficients**
We can do the same using complex coefficients. For this, we will use complex exponentials. 
!["img3"](img2.png)

```python
n = 5

expPart = np.exp(-1j*2*np.pi*t*n/T)

plt.plot(t,np.real(expPart))
plt.plot(t,np.imag(expPart))

plt.show()
```

```python
N = 50
ind = 1
c = []

for i in range(-1*N,N):
    expPart = np.exp(-1j*2*np.pi*t*i/T)
    c.append((Ts/T)*sum(np.array(f)* np.array(expPart)))
    ind+=1

x = np.arange(-1*N,N)

plt.stem(x,np.real(c),'b', markerfmt='bo')
plt.stem(x,np.imag(c),'g', markerfmt='bo')
plt.show()
```
