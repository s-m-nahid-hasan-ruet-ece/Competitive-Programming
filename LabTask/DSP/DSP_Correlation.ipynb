### **Cross Correlation**
**Definition:** The cross correlation function between two different signals is defined as the measure of similarity or coherence between one signal and the time delayed version of another signal.

The cross correlation function is defined separately for energy (or aperiodic) signals and power or periodic signals.

```python
import numpy as np
import matplotlib.pyplot as plt


y=[]
x=[]

l = int(input("Enter the lower limit of y(n) time:  "))
r = int(input("Enter the upper limit of y(n) time:  "))

for i in range(int(l),int(r)+1):
    tval = input("Enter the value of y(n) for t = %d:  " %(i))
    y.append(int(tval))


print("Time Limit for y(n): [%d, %d]"%(int(l),int(r)))
print("y(n): ",y)

y.reverse()

L = int(input("Enter the lower limit of x(n) time:  "))
R = int(input("Enter the upper limit of x(n) time:  "))

for i in range(int(L),int(R)+1):
    tval = input("Enter the value of x(n) for t = %d:  " %(i))
    x.append(int(tval))

print("Time Limit for x(n): [%d, %d]"%(int(L),int(R)))
print("x(n): ",x)
l = int(r)*-1

low = l+L
high = low + (len(x)+len(y)-2)


ht = np.linspace(int(r)*-1,int(l)*-1,(int(r)-int(l)+1))
xt = np.linspace(int(L),int(R),(int(R)-int(L)+1))
ct = np.linspace(low,high,high-low+1)

convMap =  {}
conv=[]

for i in range(0,len(y)+len(x)):
    convMap[i] = 0

for i in range(0,len(y)):
    s = 1
    for j in range(0, len(x)):        
        convMap[i+j]+= (y[i]*x[j])
    

for i in range(0,len(y)+len(x)-1):
    conv.append(convMap[i])

print("Rxy(l): ",conv)


plt.figure(figsize=(25, 8))
plt.title("Plotting of Rxy(l)")
plt.stem(ct,conv)

```

###  **Auto Correlation**
**Definition:** Autocorrelation is a mathematical representation of the degree of similarity between a given time series and a lagged version of itself over successive time intervals. It's conceptually similar to the correlation between two different time series, but autocorrelation uses the same time series twice: once in its original form and once lagged one or more time periods
