### **Linear & Non-Linear System**

**Definition of Linear System:** A system is said to be linear if it follows both the Homogenity and Additivity principles.

**Definition of Non-linear System:** A system is said to be non-linear if it does not follows the Homogenity and Additivity principles.

**Homogenity:** If the input is multiplied by a constant, the output shall also be multiplied by the same.

**Additivity:** If an input x1(t) produces output y1(t) and another input x2(t) also acting along produces output y2(t), then when both inputs acting on the system simultaneously produces output y1(t)+y2(t).

#### **Conditions**
- The output should not be zero when input applied is zero.
- Any non-linear operator can be applied on the either input or on the output to make the system non-linear.


#### **Example of Linear Systems:**
- y(n) = n*x(n)
- y(n) = x(n*n)
- y(n) = A*x(n) + B ; B=0

#### **Example of Non-linear Systems:**
- y(n) = x*x(n)
- y(n) = A*x(n) + B ; B!=0
- y(n) = e^x(n)

# **Causal & Non-causal System**

**Definition of Causal System:** A system whose output or response at any time instant (t) depends only on the present and past values of the input but not on the future values of the input is called the causal system.

**Definition of Non-causal System:** A system whose output or response at any time instant (t) depends upon future values of the input is called the non-causal system. 



#### **Example of Causal Systems:**
- y(n) = x(n) - x(n-1)
- y(n) = x(k1)+x(k2)+....+x(kn) ; where ki = [-inf,n]
- y(n) = ax(n)

#### **Example of Non-causal Systems:**
- y(n) = x(n) + 3x(n+4)
- y(n) = x(n*n)
- y(n) = x(2*n)
- y(n) = x(-n)




# **Causal System**

```python
#  Causal System
#  y(n) = x(n) - x(n-1)

n = np.linspace(1,29,30)
nn = np.linspace(1,29,28)

xx = lambda a: 3*a+5
x = [ xx(n) for n in n]

y = []

for i in range(1,len(x)-1):
    y.append(x[i]+x[i-1])

plt.figure(figsize=(25, 8))
plt.stem(nn,y)
```

```python
print("                   VALIDATION         ")

for i in range(1, len(y)):
    print("x[%d]= %lf + x[%d]=%lf  = y[%d]= %lf  " % (i,x[i],i-1,x[i-1],i,y[i-1]))
```

# **Non-causal System**

```python
#  Non-causal System
#  y(n) = x(n) + 3x(n+4)


n = np.linspace(0,29,30)
nn = np.linspace(0,25,26)

xfunc = lambda a: 3*a
x = [ xfunc(n) for n in n]

y = []

for i in range(0,len(x)-4):
    y.append(x[i] + 3*x[i+4])

plt.figure(figsize=(25, 8))
plt.stem(nn,y)
```

```python
print("                   VALIDATION         ")

for i in range(0, len(y)-4):
    print("x[%d]= %lf + 3*x[%d]=%lf  = y[%d]= %lf  " % (i,x[i],i+4,3*x[i+4],i,y[i]))
```

### **Stable & Unstable System**

**Definition of Stable System:** An arbitrary relaxed system is said to be bounded input-bounded outpout(BIBO) stable if and only if every bounded input produces a bounded output.

**Definition of Unstable System:** An arbitrary relaxed system is said to be unstable if and only if every bounded input produces an unbounded output.



#### **Example of Stable Systems:**
- y(n) = x(n) - x(n-1)
- y(n) = 5x(n)
- y(n) = x(n) + 3x(n+4)
- y(n) = x(n*n)
- y(n) = x(2*n)
- y(n) = x(-n)
#### **Example of Unstable Systems:**
- y(n) = x(k1)+x(k2)+....+x(kn) ; where ki = [-inf,n]
- y(n) = C*x(k1)+ C*x(k2)+....+Cx(kn) ; where ki = [-inf,n]




# **Stable System**

```python
#  Stable System
#  y(n) = x(2*n)

n = np.linspace(1,29,30)
nn = np.linspace(1,29,28)

xx = lambda a: a*a+3*a+50
x = [ xx(n) for n in n]

y = []

for i in range(1,len(x)-1):
    y.append(x[i]+x[i-1])

plt.figure(figsize=(25, 8))
plt.stem(nn,y)
```

# **Unstable System**

```python
#  Unstable System
#  y(n) = C*x(k1)+ C*x(k2)+....+C*x(kn) ; where ki = [0,n], C = 10

n = np.linspace(1,29,30)
C = 10

xx = lambda a: 3*a+5
x = [ xx(n) for n in n]

y = []

for i in range(0,len(x)):
    sum =0
    for j in range(0,i):
       sum+= C*x[j]
    y.append(sum)


plt.figure(figsize=(25, 8))
plt.stem(n,y)
```

### **Impulse Signal**

**Definition of Impulse Signal:** An ideal impulse signal is a signal that is zero everywhere but at the origin (t = 0), it is infinitely high. Although, the area of the impulse is finite. The unit impulse signal is the most widely used standard signal used in the analysis of signals and systems.




#### **Example of Impulse Signal:**
- (n)= { 1  when n=0 & 0 when n!=0





```python
k = input("Enter the value of k in (n+k):")
print("The value of k in (n+k): ",k)


n = np.linspace(-10-int(k),10-int(k),21)
y = []

for i in range(0,len(n)):
    if((i+int(k))<len(n) and n[i+int(k)]==0):
       y.append(1)
    else:
       y.append(0)


plt.figure(figsize=(25, 8))
plt.stem(n,y)
```

### **Resolution of Discrete Time Signal into Impulses**

```python
#  Resolution of DTS into Impulses
l = input("Enter the value of l in range [l,r]: ")
r = input("Enter the value of r in range [l,r]: ")
print("Range: [%d, %d]  "% (int(l),int(r)))


n = np.linspace(int(l),int(r),int(r)-int(l)+1)


xx = lambda a: 0.3*a+8
x = [ xx(n) for n in n]

s=""
for i in range(0,len(x)):
    ind = str(int(l)+i)
    if((int(l)+i)>0):
        s+="x("+ind+")"+"("+"n+"+ind+")"
    else:
        s+="x("+ind+")"+"("+"n"+ind+")"


    if(i!=len(x)-1):
        s+=" + "

print("Representation in Impulse: \n",s)

plt.figure(figsize=(25, 8))
plt.stem(n,x)
```
